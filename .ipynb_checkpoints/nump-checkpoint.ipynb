```python
import pandas as pd

prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]

kursUSD = 4
pricesUSD = []
for price in prices:
    pricesUSD.append(price[1] / kursUSD)

df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
df['priceUSD'] = pricesUSD
df['priceUSD'].plot(kind='line', linestyle='dashed', linewidth=2, markersize=12, grid=True, color='r', title='Price of goods (USD)')
df
```
